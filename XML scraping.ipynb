```python

def obtain_table_data(url_list, df_head):
    for i in url_list:
        print[i]
        source = urllib.request.urlopen(i)
        soup = bs.BeautifulSoup(source, 'lxml')
        table = soup.table
        table_rows = table.find_all('row')

        all_rows = []
        for row in table_rows:
            cell = row.find_all('cell')
            table_row = [i.text for i in cell]
            all_rows.append(table_row)

        df = pd.DataFrame.from_records(all_rows)
        new_header = df.iloc[0]
        df = df[1:]
        df.columns = new_header
        df = df.replace('\n', '', regex = True)
        
        df_head = df_head.append(df, ignore_index = True)
        
    return df_head
```
